## Simulacion de datos atipicos 

```python
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3-Q1

Lim_inf = Q1 - 1.5*IQR
Lim_inf

Lim_Sup = Q3 + 1.5*IQR
Lim_Sup
```

```python
df_test = df.copy()  # creacion de copia de datos limpios

indices = np.random.choice(df_test.index, 2, replace=False)
df_test.loc[indices[0],'Age'] += 3 * IQR
df_test.loc[indices[1],'Age'] += 3 * IQR


```

```python
##Diagrama de cajas y bigotes
df_test.boxplot(column='Age')
plt.show()
```

```python
##Prueba para valores atipicos

Q1_t = df_test['Age'].quantile(0.25)
Q3_T = df_test['Age'].quantile(0.75)
IQR = Q3-Q1

Lim_inf = Q1_t - 1.5*IQR
print(Lim_inf)

Lim_Sup = Q3_T + 1.5*IQR
print(Lim_Sup)
```
