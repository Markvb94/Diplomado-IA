```python
df_clean_c = df_clean[df_clean['Cultivar'] == 'Caly'] ## seleccion de valores especificos
df_clean_c
```

```python
df_clean_c['Set'].unique()
```

```python
df_test = df_clean_c[df_clean_c['Set'] == 'Val Ext'] ## creación de set de prueba
df_cal = df_clean_c[df_clean_c['Set'] != 'Val Ext'] ## creación de set de calibracion
```

```python
y_test = df_test['DM']
y_cal = df_cal['DM']
```

```python
df_cal.head()
```

```python
df_test.head()
```

```python
x_cal = df_cal.iloc[:,9:] ## obtencion de valores de x 
x_test = df_test.iloc[:,9:]
x_test
```

### MLR

```python
##librerias

import pandas as pd
import numpy as np
import matplotlib as plt
from sklearn.decomposition import PCA 
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
PCR_pipeline = Pipeline([
    ('Scalar',StandardScaler()),
    ('pca',PCA()),
    ('regresion',LinearRegression())
 ])

MLR_pipeline = Pipeline([
    ('Scalar',StandardScaler()),
    ('regresion',LinearRegression())
 ])
```

```python
PCR_pipeline.fit(x_cal,y_cal)
```

```python
MLR_pipeline.fit(x_cal,y_cal)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
## Metricas regresión lineal multiple
pred_mlr = MLR_pipeline.predict(x_test)
r2_score(y_test,pred_mlr)
np.square(mean_squared_error(y_test,pred_mlr))
```

```python
## Metricas pcr
pred_pcr = PCR_pipeline.predict(x_test)
print(r2_score(y_test,pred_pcr))
print(np.sqrt(mean_squared_error(y_test,pred_mlr)))
```

```python
pca = PCR_pipeline.named_steps['pca'] ## calculo de componentes
explained_variance = pd.DataFrame(pca.explained_variance_ratio_)
cumsum = explained_variance.cumsum()
cumsum.plot()
```

```python
### busqueda por rejilla

from sklearn.model_selection import GridSearchCV,KFold

##pca_params={
##    'pca__n_components': [6,34,100,150,280] ### rango de evaliacion
##}
pca_params={
    'pca__n_components':np.arange(1,280) ### rango de evaliacion
}



cv= KFold(n_splits=10,shuffle=True)
pca_grid = GridSearchCV(PCR_pipeline,pca_params,cv = cv,scoring='neg_mean_squared_error')
pca_grid.fit(x_cal,y_cal)
```
