```python
#rafico de dispersion
import numpy as np
x = np.random.rand(100)
y= np.random.rand(100)
plt.scatter(x,y)
```

```python
## histograma
x= np.random.normal(100)
plt.hist(x,bins=10)
```

```python
##grafico de bigotes

datos = [np.random.normal(0,std,100) for std in range (1,4)]
plt.boxplot(datos,vert = True)
plt.show()
```

```python
##serie de tiempo
import pandas as pd

data_range = pd.date_range(start='20230101',end='20231231',freq='d')
precios = np.cumsum(np.random.randn(len(data_range)))
```

```python
data =  pd.DataFrame({'Date':data_range,'Precios':precios})
data.head()
```

```python
serie_tiempo =plt.plot(data['Date'],data['Precios'],label = 'Precio')
```

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)

categorias = ['a','b','c','d']
valores = [5,6,7,8]

w= np.random.rand(50)
z= np.random.rand(50)
```

```python
##hacer 4 gifuras en misma hoja

fig, axs = plt.subplots(2,2, figsize=(15,10))
axs[0,0].plot(x,y1)
axs[0,0].set_title('Grafico de lineas')

axs[0,1].bar(categorias,valores)
axs[0,1].set_title('Grafico de barras')

axs[1,0].plot(data['Date'],data['Precios'],label = 'Precio')
axs[1,0].set_title('Grafico series de tiempo')

axs[1,1].boxplot(datos,vert = False)
axs[1,1].set_title('Grafico de caja/bigotes')


plt.show()
```
