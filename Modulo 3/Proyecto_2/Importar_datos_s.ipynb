```python
carpeta = r'./Original_dataset'
```

```python
dataframe = []

##ciclo para traer todos los archivos
for archivo in os.listdir(carpeta):
    df = pd.read_csv(carpeta + '/' + archivo)
    columnas_primarias = ['Fecha','Hora']
    zonas = ['Noreste','Noroeste','Centro','Suroeste','Sureste']
    for zonas in zonas:
        columnas_primarias.append(f'{zonas} PM10')
        columnas_primarias.append(f'{zonas} monóxido de carbono')
    df_interes = df[columnas_primarias]
    dataframe.append(df_interes)

df_total = pd.concat(dataframe,ignore_index= True)
```

```python
## reemplazo de valores

import numpy as np

df_total.replace(-99,np.nan,inplace=True)

```

```python
def plot_time_series(df, variable, zones, ylabel, title, start_date=None, end_date=None):
    plt.figure(figsize=(15, 5))
    if start_date and end_date:
        df = df.loc[start_date:end_date]
    
    for zone in zones:
        column = f"{zone}-{variable}"
        if column in df.columns:
            plt.plot(df.index, df[column], label=column)
        else:
            print(f"Columna {column} no encontrada en el DataFrame")
    plt.xlabel('Fecha_Hora')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend()
    plt.show()
```
